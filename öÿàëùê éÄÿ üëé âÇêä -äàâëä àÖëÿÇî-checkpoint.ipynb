```python
#removing unnessery data

Species.drop(["SPEC","IUCN Category", "COU", "Unit", "Unit Code", "PowerCode Code",
              "PowerCode", "Reference Period Code", "Reference Period", "Flag Codes"
              , "Flags"], axis = 1, inplace = True) 
```

```python
#changeing all valuse to %
medals['CRITICALAmphibians']=(medals['CRITICALAmphibians']*100/medals['TOT_KNOWNAmphibians'])
medals['ENDANGEREDAmphibians']=(medals['ENDANGEREDAmphibians']*100/medals['TOT_KNOWNAmphibians'])
medals['VULNERABLEAmphibians']=(medals['VULNERABLEAmphibians']*100/medals['TOT_KNOWNAmphibians'])
medals['CRITICALBirds']=(medals['CRITICALBirds']*100/medals['TOT_KNOWNBirds'])
medals['ENDANGEREDBirds']=(medals['ENDANGEREDBirds']*100/medals['TOT_KNOWNBirds'])
medals['VULNERABLEBirds']=(medals['VULNERABLEBirds']*100/medals['TOT_KNOWNBirds'])
medals['CRITICALFish']=(medals['CRITICALFish']*100/medals['TOT_KNOWNFish'])
medals['ENDANGEREDFish']=(medals['ENDANGEREDFish']*100/medals['TOT_KNOWNFish'])
medals['VULNERABLEFish']=(medals['VULNERABLEFish']*100/medals['TOT_KNOWNFish'])
medals['CRITICALFreshwaterFish']=(medals['CRITICALFreshwaterFish']*100/medals['TOT_KNOWNFreshwaterFish'])
medals['ENDANGEREDFreshwaterFish']=(medals['ENDANGEREDFreshwaterFish']*100/medals['TOT_KNOWNFreshwaterFish'])
medals['VULNERABLEFreshwaterFish']=(medals['VULNERABLEFreshwaterFish']*100/medals['TOT_KNOWNFreshwaterFish'])
medals['CRITICALInvertebrates']=(medals['CRITICALInvertebrates']*100/medals['TOT_KNOWNInvertebrates'])
medals['ENDANGEREDInvertebrates']=(medals['ENDANGEREDFreshwaterFish']*100/medals['TOT_KNOWNInvertebrates'])
medals['VULNERABLEInvertebrates']=(medals['VULNERABLEFreshwaterFish']*100/medals['TOT_KNOWNInvertebrates'])
medals['CRITICALLichens']=(medals['CRITICALLichens']*100/medals['TOT_KNOWNLichens'])
medals['ENDANGEREDLichens']=(medals['ENDANGEREDLichens']*100/medals['TOT_KNOWNLichens'])
medals['VULNERABLELichens']=(medals['VULNERABLELichens']*100/medals['TOT_KNOWNLichens'])
medals['CRITICALMammals']=(medals['CRITICALMammals']*100/medals['TOT_KNOWNMammals'])
medals['ENDANGEREDMammals']=(medals['ENDANGEREDMammals']*100/medals['TOT_KNOWNMammals'])
medals['VULNERABLEMammals']=(medals['VULNERABLEMammals']*100/medals['TOT_KNOWNMammals'])
medals['CRITICALMarineFish']=(medals['CRITICALMarineFish']*100/medals['TOT_KNOWNMarineFish'])
medals['ENDANGEREDMarineFish']=(medals['ENDANGEREDMarineFish']*100/medals['TOT_KNOWNMarineFish'])
medals['VULNERABLEMarineFish']=(medals['VULNERABLEMammals']*100/medals['TOT_KNOWNMarineFish'])
medals['CRITICALMosses']=(medals['CRITICALMosses']*100/medals['TOT_KNOWNMosses'])
medals['ENDANGEREDMosses']=(medals['ENDANGEREDMosses']*100/medals['TOT_KNOWNMosses'])
medals['VULNERABLEMosses']=(medals['VULNERABLEMosses']*100/medals['TOT_KNOWNMosses']) 
medals['CRITICALReptiles']=(medals['CRITICALReptiles']*100/medals['TOT_KNOWNReptiles'])
medals['ENDANGEREDReptiles']=(medals['ENDANGEREDReptiles']*100/medals['TOT_KNOWNReptiles'])
medals['VULNERABLEReptiles']=(medals['VULNERABLEReptiles']*100/medals['TOT_KNOWNReptiles'])                              
medals['CRITICALVascularplants']=(medals['CRITICALVascularplants']*100/medals['TOT_KNOWNVascularplants'])
medals['ENDANGEREDVascularplants']=(medals['ENDANGEREDVascularplants']*100/medals['TOT_KNOWNVascularplants'])
medals['VULNERABLEVascularplants']=(medals['VULNERABLEVascularplants']*100/medals['TOT_KNOWNVascularplants'])                  

```

# Selecting the features for the model
<a id='Selecting'></a>


**Fish**

```python
#our model
linregmodel=smf.ols(formula='CRITICALFish~PrimaryT2015+PrimaryT2016+PrimaryT2017+SecondaryT2015+SecondaryT2016+SecondaryT2017+TertiaryT2015+TertiaryT2016+TertiaryT2017',data=final)
result=linregmodel.fit()
y=final.iloc[:,1:2]
result.summary()
```

# Split the data & Train the model
<a id='split'></a>
<a id='train'></a>

```python
x=final.iloc[:,28:37]

# split to train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)
# now generate predictions
y_train_predicted = result.predict(x_train)
y_test_predicted = result.predict(x_test)
```

# MSE RMSE MAE
<a id='mse'></a>

```python
from sklearn import metrics
print("------ TRAIN DATA-CRITICALFish ------")
print("MSE:",metrics.mean_squared_error(y_train, y_train_predicted))
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_train, y_train_predicted)))
print("MAE:",metrics.mean_absolute_error(y_train, y_train_predicted))
```

```python
print("------ TEST DATA ------")
print("MSE:",metrics.mean_squared_error(y_test, y_test_predicted))
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_test, y_test_predicted)))
print("MAE:",metrics.mean_absolute_error(y_test, y_test_predicted))
```
